***Estimation*** is an inverse problem with the goal of recovering the underlying parameter θ of a distribution \f_X(x;θ) based on observed samples X1,X2,...


```python
# Python code to visualize a graph
import networkx as nx
import numpy as np
```

```python
n = 40  # Number of nodes
p = 0.3   # probability
m = np.round(((n ** 2)/2)*p)   # Number of edges
G = nx.gnm_random_graph(n,m)  # Graph
A = nx.adjacency_matrix(G)  # Adj matrix
nx.draw(G)  # Drawing
p_ML = np.mean(A)   # ML estimate
```
